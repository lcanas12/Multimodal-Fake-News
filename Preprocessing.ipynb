```python
def sendRequest(url):
    try:
        page = requests.get(url, timeout=10)
        size = page.headers['Content-length']
        if int(size) > 5000:
            return url
        
    except Exception as e:
        #print("error:", e)
        return url.replace(url, "")

    # check status code
    if (page.status_code != 200):
        return url.replace(url, "")
    
    return url.replace(url, "")
```

```python
#regular expression to filter items in the images column that starts with http but not end with .gif nor .ico    
def clean_img1(url:str):
    re_htttp_not_gif = re.compile(r'^http(?!.*?\.gif$).*$')    
    if re_htttp_not_gif.match(url) is None:
        return url.replace(url, "")
    else: return url
def clean_img2(url:str):
    re_not_icon = re.compile(r'(?!.*?\.ico$).*$')
    if re_not_icon.match(url) is None:
        return url.replace(url, "")
    else: return url
def clean_img3(url:str):
    re_not_icon = re.compile(r'^(?!.*\bicon\b).*$')
    if re_not_icon.match(url) is None:
        return url.replace(url, "")
    else: return url
```

```python
def clean_text(text):
    '''Remove links,remove tabs
    and remove unwanted spacing.'''
    text = re.sub('https?://\S+|www\.\S+', '', text) 
    text = re.sub('\n', ' ', text)
    text = re.sub('\s{2}', ' ', text)
    return text
```
